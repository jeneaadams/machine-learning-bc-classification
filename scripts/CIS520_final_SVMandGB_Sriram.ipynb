# Support Vector Machine

```python
C_options = [0.005, 0.01, 0.1, 1, 1.25, 1.5, 1.75, 2]
kernel_options = ['linear', 'poly', 'rbf', 'sigmoid']

accuracies = [0, 0, 0, 0, 0]
bestCandKernelCombo = [[0,''], [0,''], [0,''], [0,''], [0,'']]
```

```python
bestCombo = []
bestAccuracyScore = -1
for kernelValue in kernel_options:
    for cValue in C_options:
        currentModel = SVC(C = cValue, kernel = kernelValue)
        currentModel.fit(Xtrain_1, ytrain_1)
        yValPred = currentModel.predict(Xval_1)
        currentAccuracyScore = accuracy_score(yval_1, yValPred)
        if(currentAccuracyScore > bestAccuracyScore):
            bestCombo = [cValue, kernelValue]
            bestAccuracyScore = currentAccuracyScore

accuracies[0] = bestAccuracyScore
bestCandKernelCombo[0] = bestCombo
```
